```python
%pip install gymnasium
import gymnasium as gym
import matplotlib.pyplot as plt
```

```python
#creating Environment
env = gym.make('MountainCar-v0', render_mode='rgb_array')
#making an state
initial_state,info=env.reset(seed=42)
position = initial_state[0]
velocity = initial_state[1]

print(f"The position of the car along the x-axis is {position} (m)")
print(f"The velocity of the car is {velocity} (m/s)")
```

```python
#creating a function , for showing the current state of the cartPole
def render():
    state_image = env.render()
    plt.imshow(state_image)
    plt.show()
```

```python
render()
```

```python
#here action= 1 move right
#and action = 0 move left

action = 1
state, reward, terminated,_,_ = env.step(action)
print(state)
print(reward)
print(terminated)
```

```python
while(not terminated):
    action = 1
    state, reward, terminated,_,_ = env.step(action)
    # state = next state of the environment
    #reward = Current State
    # terminated = Is the operation is terminated / not
    render()
```
