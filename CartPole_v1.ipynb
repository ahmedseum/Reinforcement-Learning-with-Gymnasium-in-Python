```python
%pip install gymnasium
import gymnasium as gym
import matplotlib.pyplot as plt
```

```python
#creating Environment
env = gym.make('CartPole-v1', render_mode='rgb_array')
#making an state
state,info=env.reset(seed=42)
print(state)
```

```python
#creating a function , for showing the current state of the cartPole
def render():
    plt.imshow(env.render())
    plt.show()
```

```python
render()
```

```python
#here action= 1 move right
#and action = 0 move left

action = 1
state, reward, terminated,_,_ = env.step(action)
print(state)
print(reward)
print(terminated)
```

```python
while(not terminated):
    action = 1
    state, reward, terminated,_,_ = env.step(action)
    # state = next state of the environment
    #reward = Current State
    # terminated = Is the operation is terminated / not
    render()
```
